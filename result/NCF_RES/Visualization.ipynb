```python
import pandas as pd
nmf = pd.read_csv('nmf_train_val_res.csv')
nmf['epoch'] = [i for i in range(len(nmf))]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.plot(nmf['train_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('NMF TRAIN LOSS')
```

```python
plt.plot(nmf['valid_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('NMF VALID LOSS')
```

```python
plt.plot(nmf['HR'])
plt.xlabel('epoch')
plt.ylabel('HR')
plt.title('NMF HR')
```

```python
plt.plot(nmf['NDCG'])
plt.xlabel('epoch')
plt.ylabel('NDCG')
plt.title('NMF NDCG')
```

```python
plt.plot(nmf['epoch'], nmf['train_loss'],label = 'train')
plt.plot(nmf['epoch'], nmf['valid_loss'],label = 'valid')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('NMF train&val loss')
```

```python
HR = pd.DataFrame({'Epoch':mlp['Epoch'],'NMF_HR':nmf['HR'],'GMF_HR':gmf['HR'],'MLP_HR':mlp['HR']})
plt.plot(HR['Epoch'], HR['NMF_HR'],label = 'NMF')
plt.plot(HR['Epoch'], HR['GMF_HR'],label = 'GMF')
plt.plot(HR['Epoch'], HR['MLP_HR'],label = 'MLP')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('HR')
plt.title('MODEL HR')
```
