## 평점만 고려한 추천시스템-NDCG : 0.63

```python
def recommend(mbrNo, df, top=10):
    # 사용자가 평가한 상품의 점수를 가져옴
    user_ratings = df[df['mbrNo'] == mbrNo]['gdasScrVal']

    # 사용자가 평가한 상품을 제외한 모든 상품과의 코사인 유사도 계산
    cosine_sim = cosine_similarity(df['gdasScrVal'].values.reshape(-1, 1), df['gdasScrVal'].values.reshape(-1, 1))

    # 사용자가 평가한 상품 인덱스 추출
    user_idx = df[df['mbrNo'] == mbrNo].index.values[0]

    # 코사인 유사도가 가장 높은 상위 상품 추출
    sim_scores = cosine_sim[user_idx]
    top_similar_indices = sim_scores.argsort()[::-1][:top]
    top_similar_goods = df.iloc[top_similar_indices]['goodsNm'].tolist()

    # 중복된 상품 제거
    top_similar_goods = list(set(top_similar_goods))
    # 추천 상품 목록 반환
    return top_similar_goods
```

```python
def print_recommendations(user_id, recommendations):
    print(f"Recommended goods for User {user_id}:")
    for i, goods in enumerate(recommendations):
        print(f"{i+1}. {goods}")

# 사용자 아이디를 입력하여 추천 상품을 출력
user_id = "M0000000004414"
recommendations = recommend(user_id, df, top=10)
print_recommendations(user_id, recommendations)
```

```python
from sklearn.metrics import ndcg_score
# 상품 이름을 고유한 ID로 매핑하는 사전 생성
item_id_dict = {item: i for i, item in enumerate(new_temp['goodsNm'].unique())}


def evaluate_recommendation(test_items, recommended_items, top_k):
    # NDCG 계산
    def ndcg(recommended, relevant):
        dcg = 0.0
        idcg = 0.0
        for i in range(len(recommended)):
            if recommended[i] in relevant:
                dcg += 1.0 / np.log2(i + 2)
        for i in range(len(relevant)):
            idcg += 1.0 / np.log2(i + 2)
        return dcg / idcg

    # Hit Rate 계산
    def hit_rate(recommended, relevant):
        for item in recommended:
            if item in relevant:
                return 1
        return 0

    # 추천된 상위 k개의 아이템 추출
    top_k_recommended = recommended_items[:top_k]

    # NDCG와 Hit Rate 계산
    ndcg_score = ndcg(top_k_recommended, test_items)
    hit_rate_score = hit_rate(top_k_recommended, test_items)

    return ndcg_score, hit_rate_score
```

```python
def recommend2(mbrNo, df, top=10):
    # 사용자가 평가한 상품의 등급을 가져옴
    user_ratings = df[df['mbrNo'] == mbrNo]['gdasScrVal']

    # 리뷰 본문 벡터화 (CountVectorizer 또는 TF-IDFVectorizer 사용)
    review_vectorizer = CountVectorizer()
    review_X = review_vectorizer.fit_transform(df['kor'])

    # 키워드 벡터화 (CountVectorizer 또는 TF-IDFVectorizer 사용)
    keyword_vectorizer = CountVectorizer()
    keyword_X = keyword_vectorizer.fit_transform(df['keyword'])

    # 사용자가 평가한 상품 인덱스 추출
    user_idx = df[df['mbrNo'] == mbrNo].index.values[0]

    # 리뷰 본문의 코사인 유사도 계산
    review_cosine_sim = cosine_similarity(review_X)
    review_sim_scores = review_cosine_sim[user_idx]
    review_top_similar_indices = review_sim_scores.argsort()[::-1][:top]
    review_top_similar_goods = df.iloc[review_top_similar_indices]['goodsNm'].tolist()

    # 키워드의 코사인 유사도 계산
    keyword_cosine_sim = cosine_similarity(keyword_X)
    keyword_sim_scores = keyword_cosine_sim[user_idx]
    keyword_top_similar_indices = keyword_sim_scores.argsort()[::-1][:top]
    keyword_top_similar_goods = df.iloc[keyword_top_similar_indices]['goodsNm'].tolist()

    # 중복된 상품 제거
    top_similar_goods = list(set(review_top_similar_goods + keyword_top_similar_goods))

    # 추천 상품 목록 반환
    return top_similar_goods
```

```python
def recommend3(user_id, df, top=10):
    # 해당 사용자의 평가 상품이 있는지 확인
    if user_id not in df['mbrNo'].values:
        print(f"No ratings found for User {user_id}")
        return []

    # 사용자가 평가한 상품 인덱스 추출
    user_indices = df[df['mbrNo'] == user_id].index

    # 리뷰 본문 벡터화 (CountVectorizer 또는 TF-IDFVectorizer 사용)
    review_vectorizer = CountVectorizer()
    review_X = review_vectorizer.fit_transform(df['kor'])

    # 리뷰 본문의 코사인 유사도 계산
    review_cosine_sim = cosine_similarity(review_X)

    # 유사한 상품 추천
    similar_goods_indices = []
    for user_idx in user_indices:
        sim_scores = review_cosine_sim[user_idx]
        top_similar_indices = sim_scores.argsort()[::-1][1:top+1]  # 자기 자신은 제외
        similar_goods_indices.extend(top_similar_indices)

    # 추천 상품 목록 반환 (중복 제거)
    recommendations = list(set(df.loc[similar_goods_indices]['goodsNm'].tolist()))
    recommendations.sort()  # 알파벳 순으로 정렬
    return recommendations
```

```python
def recommend5(mbrNo, df, top=10):
    # 키워드 벡터화 (CountVectorizer 또는 TF-IDFVectorizer 사용)
    keyword_vectorizer = CountVectorizer()
    keyword_X = keyword_vectorizer.fit_transform(df['keyword'])

    # 사용자가 평가한 상품 인덱스 추출
    user_idx = df[df['mbrNo'] == mbrNo].index.values[0]

    # 키워드의 코사인 유사도 계산
    keyword_cosine_sim = cosine_similarity(keyword_X)
    keyword_sim_scores = keyword_cosine_sim[user_idx]
    keyword_top_similar_indices = keyword_sim_scores.argsort()[::-1][:top]
    keyword_top_similar_goods = df.iloc[keyword_top_similar_indices]['goodsNm'].tolist()

    # 중복된 상품 제거
    keyword_top_similar_goods = list(set(keyword_top_similar_goods))

    # 추천 상품 목록 반환
    return keyword_top_similar_goods
```
